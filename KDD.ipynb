```python
df = df.rename(columns={"Year of release": "Year"}) #원하는 컬럼 이름으로 변경하기 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(150,10)) #(가로,세로)
sns.countplot(x="Genre", data=df, order=df["Genre"].value_counts().index)
plt.xticks(rotation=45)
plt.title("K-Drama 장르별 개수")
plt.show()
```

```python
# 2. 연도별 드라마 개수 분석
plt.figure(figsize=(10,5))
sns.countplot(x="Year", data=df, order=sorted(df["Year"].unique()))
plt.xticks(rotation=45)
plt.title("연도별 K-Drama 개수")
plt.show()
```

```python
# 3. 평균 평점이 높은 드라마 Top 10
top_drama = df.sort_values(by="Rating", ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x="Rating", y="Title", data=top_drama)
plt.title("평점이 높은 K-Drama TOP 10")
plt.show()
```
